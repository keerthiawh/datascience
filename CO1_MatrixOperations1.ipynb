```python
#Matrix Operations

import numpy as np
a = np.array([[1,2,3],[1,2,3],[1,2,3]])
b = np.array([[7,8,9],[7,8,9],[7,8,9]])
print("Addition of 2 matrix\n")
print(np.add(a,b))
print(a+b)

print("Subtraction of 2 matrix\n")
print(np.subtract(b,a))
print(b-a)

print("Multiplication of 2 matrix\n")
print(np.multiply(a,b))
print(a*b)

print("Division of 2 matrix\n")
print(np.divide(b,a))
print(b/a)
```

```python
print("Product Matrix\n",np.dot(a,b))
print("Transpose of First matrix\n",np.transpose(b))
```

The Singular Value Decomposition (SVD) of a matrix is a factorization of that matrix into three matrices.

The SVD of a matrix X of dimension n×d is given by:
svd.webp

U and V are square orthogonal:
D is diagonal of dimension d×n


SVD-1.jpg.webp

```python

#single value decomposition
import numpy as np

#Creating a matrix A
A = np.array([[3,4,3],[1,2,3],[4,2,1]])

#Performing SVD
U, D, VT = np.linalg.svd(A)
print("\nU\n",U)
print("\nD\n",D)
print("\nVT\n",VT)
#Checking if we can remake the original matrix using U,D,VT
A_remake = (U @ np.diag(D) @ VT)
print("\nREMAKE\n",A_remake)

```
