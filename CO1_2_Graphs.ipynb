***scatter plot***

x_axis_data- An array containing x-axis data

y_axis_data- An array containing y-axis data

s- marker size (can be scalar or array of size equal to size of x or y)

c- color of sequence of colors for markers

marker- marker style

cmap- cmap name

linewidths- width of marker border

edgecolor- marker border color

```python

import numpy as np
import matplotlib.pyplot as plt
x=np.array([0,1,2,3,4,5,6])
y=np.array([0,1,4,9,16,25,36])
plt.scatter(x,y,c='red',marker='*',s= 200,edgecolors='blue')
x=np.array([0,1,2,3,4,5,6])
y=np.array([0,1,8,27,64,75,216])
plt.scatter(x,y,c='yellow',marker='*',s=100,edgecolors='blue')

plt.legend(['Square','Cube'])
plt.show()
```

**histogram**

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

histtype = {‘bar’, ‘barstacked’, ‘step’, ‘stepfilled’}

align : This parameter is an optional parameter and it controls how the histogram is plotted. {‘left’, ‘mid’, ‘right’}


```python
import numpy as np
x = np.random.normal(170, 2, 200)           #loc scale  size
#print(x)                                   #loc = Mean (“centre”) of the distribution,
                                           #scale = Standard deviation (spread or “width”) of the distribution. Must be non-negative.
                                           #size = int or tuple of ints, optional
plt.hist(x,histtype='bar', align='right', orientation='vertical',color='violet')
#plt.xlabel("Height")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating x axis with range and y axis with Sine
# Function for Plotting Cosine Graph
x = np.arange(0, 30, 0.1)
y = np.sin(x)

# Plotting coine Graph
plt.plot(x, y, color='green')
x = np.arange(0,30, 0.1)
y = np.cos(x)
plt.plot(x, y, color='blue')
plt.legend(['Sin(x)','Cos(x)'],loc='upper right')
plt.show()
```

***subplot***

plt.subplot(1, 2, 1)

The figure has 1 row, 2 columns, and this plot is the first plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 30, 0.1)
y = np.sin(x)
plt.subplot(1,3,1)
plt.plot(x, y, color='green')
plt.legend(['Sin(x)'],loc='upper right')

x = np.arange(0,30, 0.1)
y = np.cos(x)
plt.subplot(1,3,2)
plt.plot(x, y, color='blue')
plt.legend(['Cos(x)'],loc='lower right')

x = np.arange(0,30, 0.1)
y = np.tan(x)
plt.subplot(1,3,3)
plt.plot(x, y, color='red')
plt.legend(['Tan(x)'],loc='upper left')


plt.show()
```
